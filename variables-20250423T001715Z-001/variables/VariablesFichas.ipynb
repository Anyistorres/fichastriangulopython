```python
#DECLARACIÃN DE VARIABLES DE DIFERENTES TIPOS E IMPRESIÃN EN PYTHON

#ENTEROS (INT):
edad = 26
print("Edad:", edad)                      # Salida: Edad: 26
```

```python
#REAL (FLOAT):
estatura = 1.68
print("Estatura:", estatura )                  # Salida: Estatura: 1.68
```

```python
#CADENAS DE TEXTO O CARACTER (STR)
nombre = "Anyis Torres"
print("Nombre:", nombre)                  # Salida: Nombre: Anyis Torres
```

```python
#BOOLEANOS (BOOL)
es_activo = True
tiene_descuento = False
print("Â¿EstÃ¡ activo?", es_activo)         # Salida: Â¿EstÃ¡ activo? True
print("Â¿Tiene descuento?", tiene_descuento) # Salida: Â¿Tiene descuento? False
```

```python
#LISTAS (LIST)
frutas = ["manzana", "banana", "naranja"]
numeros = [1, 2, 3, 4, 5]
print("Frutas:", frutas)                  # Salida: Frutas: ['manzana', 'banana', 'naranja']
print("NÃºmeros:", numeros)                # Salida: NÃºmeros: [1, 2, 3, 4, 5]
```

```python
#DICCIONARIOS (DICT)
persona = {
    "nombre": "Anyis",
    "edad": 26,
    "ciudad": "Arauca"
}
print("Datos de persona:", persona)       # Salida: Datos de persona: {'nombre': 'Anyis', 'edad': 26, 'ciudad': 'Arauca'}
```
