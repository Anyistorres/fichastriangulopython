**LOGICA COMPUTACIONAL:** Es la base del razonamiento en programación y algoritmos. Se encarga de evaluar condiciones y tomar decisiones basadas en valores booleanos (True o False). Sus elementos principales son:


- Operadores de comparación: generan valores booleanos.

Ejemplo: 5 > 3 (True), 2 == 0 (False).

- Operadores lógicos: Operadores que combinan/niegan proposiciones (and, or, not, etc.).

**OPERADOR DE COMPARACIÓN (RELACIONALES):** Son símbolos que comparan dos valores y devuelven un resultado booleano: True (verdadero) o False (falso).

**OPERTADOR LOGICO EN PYTHON:** Los operadores lógicos en Python se utilizan para combinar o modificar expresiones booleanas (True o False). Permiten tomar decisiones basadas en múltiples condiciones.

**Tipos de operadores lógicos:**

- **and (Y lógico):** Devuelve True si todas las condiciones son verdaderas.
- **or (O lógico):** Devuelve True si al menos una condición es verdadera.
- **not (Negación):** Invierte el valor booleano (True → False, False → True).

Captura de pantalla 2025-04-21 235616.png

```python
#EJEMPLO:
# Definimos las condiciones
esta_soleado = True
tenemos_tiempo = False

# Usamos operadores lógicos
salir_a_pasear = esta_soleado and tenemos_tiempo  # ¿Ambos True?
salir_con_paraguas = esta_soleado or tenemos_tiempo  # ¿Al menos uno True?
quedarse_en_casa = not esta_soleado  # ¿NO está soleado?

# Mostramos resultados
print("¿Salir a pasear?", salir_a_pasear)       # False (porque no hay tiempo)
print("¿Salir con paraguas?", salir_con_paraguas) # True (porque está soleado)
print("¿Quedarse en casa?", quedarse_en_casa)    # False (porque SÍ está soleado)
```

**NOTAS:**
- Los operadores lógicos siempre devuelven True o False.
- Pueden combinarse con operadores de comparación (>, <, ==, etc.).
